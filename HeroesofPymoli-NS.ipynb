```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
data_file = pd.read_csv(file_to_load)
```

```python
# Display df

data_file.head()
```

```python
# Total Players

totalplayers = data_file["SN"].nunique()
totalplayers
```

```python
# Gender Demographics

uniquecount = data_file["Item ID"].nunique()
avgprice = data_file["Price"].mean()
purchasecount = data_file["Purchase ID"].count()
revenue = data_file["Price"].sum()

summary_df = pd.DataFrame({
    "# Unique Items":[uniquecount],
    "Average Price":[avgprice],
    "# Purchases":[purchasecount],
    "TTL Revenue":[revenue]
})

summary_df = summary_df.round(2)
summary_df
```

```python
# Gender Demographics

# count of each gender
playerdemo = data_file[["SN","Gender"]]
playerdemo = playerdemo.drop_duplicates()
gendercount = playerdemo["Gender"].value_counts()
genderper = (gendercount/totalplayers)*100

gender_df = pd.DataFrame({
    "Gender Count" : gendercount,
    "Percentage" : genderper
    
})
gender_df = gender_df.round(2)
gender_df
```

```python
# Purchase Analysis by gender

# purchasebygender = data_file["Purchase ID"].count()
purchasecount = data_file.groupby(["Gender"]).count()["Price"]

avgprice = data_file.groupby(["Gender"]).mean()["Price"]
avgprice = avgprice.round(2)

totalvalue = data_file.groupby(["Gender"]).sum()["Price"]

avgpur = totalvalue/purchasecount

#playerdemo, gender_df

purchasebygender = pd.DataFrame({
    "Purchase Count" : purchasecount,
    "Average Purchase Price" : avgprice,
    "Total Value" : totalvalue,
    "Avg Total Purchase Per Person" : totalvalue/purchasecount
})

purchasebygender.head()
```
